## Data preprocessing

```python
# Drop PassengerId
train_df.drop(columns=['PassengerId'], inplace=True)
```

```python
# Checking Null values
train_df.isnull().sum()
```

```python
# Cabin column having more than 70% NaN value
train_df.drop(columns=['Cabin'],inplace=True)
```

```python
print(train_df['Embarked'].value_counts())

# So S is more appeared value in Embarked column
train_df['Embarked'].fillna('S', inplace=True)
```

```python
# Imputting missing value for Age
train_df['Age'].fillna(train_df['Age'].mean(), inplace=True)
```

```python
# categorical data
train_df['Sex'] = pd.get_dummies(train_df['Sex'], drop_first=True)
```

```python
# checking duplicate
train_df.duplicated().value_counts()
```
